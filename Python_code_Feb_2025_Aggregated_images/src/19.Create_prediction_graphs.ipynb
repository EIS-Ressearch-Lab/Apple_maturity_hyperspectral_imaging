# Generate prediction vs actual graphs for the three features, merge them by run number

```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data

CNN2D_brix_act = pickle.load(open('/media/2tbdisk3/data/Haidee/Results/Predictions/run1/run1_Actual_brix_train.pkl', 'rb'))
CNN2D_brix_pred = pickle.load(open('/media/2tbdisk3/data/Haidee/Results/Predictions/run1/run1_Predicted_Brix_train.pkl', 'rb'))
CNN2D_firm_act = pickle.load(open('/media/2tbdisk3/data/Haidee/Results/Predictions/run1/run1_Actual_firmness_train.pkl', 'rb'))
CNN2D_firm_pred = pickle.load(open('/media/2tbdisk3/data/Haidee/Results/Predictions/run1/run1_Predicted_Firmness_train.pkl', 'rb'))
CNN2D_starch_act = pickle.load(open('/media/2tbdisk3/data/Haidee/Results/Predictions/run1/run1_Actual_starch_train.pkl', 'rb'))
CNN2D_starch_pred = pickle.load(open('/media/2tbdisk3/data/Haidee/Results/Predictions/run1/run1_Predicted_Starch_train.pkl', 'rb'))

CNN3D_brix = pd.read_csv('/media/2tbdisk3/data/Haidee/Results/Predictions/run_9/brix_prediction_results.csv')
CNN3D_firm = pd.read_csv('/media/2tbdisk3/data/Haidee/Results/Predictions/run_9/firmness_prediction_results.csv')
CNN3D_starch = pd.read_csv('/media/2tbdisk3/data/Haidee/Results/Predictions/run_9/starch_prediction_results.csv')

hybrid_brix = pd.read_csv('/media/2tbdisk3/data/Haidee/Results/Predictions/run_10_hybrid_results/brix_hybrid_prediction_results.csv')
hybrid_firm = pd.read_csv('/media/2tbdisk3/data/Haidee/Results/Predictions/run_10_hybrid_results/firmness_hybrid_prediction_results.csv')
hybrid_starch = pd.read_csv('/media/2tbdisk3/data/Haidee/Results/Predictions/run_10_hybrid_results/starch_hybrid_prediction_results.csv')


ViT_brix = pd.read_csv('/media/2tbdisk3/data/Haidee/Results/Predictions/run_11_ViT_results/brix_vit_prediction_results.csv')
ViT_firm = pd.read_csv('/media/2tbdisk3/data/Haidee/Results/Predictions/run_11_ViT_results/firmness_vit_prediction_results.csv')
ViT_starch = pd.read_csv('/media/2tbdisk3/data/Haidee/Results/Predictions/run_11_ViT_results/starch_vit_prediction_results.csv')

ViT_opt_brix = pd.read_csv('/media/2tbdisk3/data/Haidee/Results/Predictions/run_20_ViT_results/brix_vit_prediction_results.csv')
ViT_opt_firm = pd.read_csv('/media/2tbdisk3/data/Haidee/Results/Predictions/run_20_ViT_results/firmness_vit_prediction_results.csv')
ViT_opt_starch = pd.read_csv('/media/2tbdisk3/data/Haidee/Results/Predictions/run_20_ViT_results/starch_vit_prediction_results.csv')
```

```python

# Flatten in case they are 2D arrays
CNN2D_brix_act = np.array(CNN2D_brix_act).flatten().tolist()
CNN2D_brix_pred = np.array(CNN2D_brix_pred).flatten().tolist()
CNN2D_firm_act = np.array(CNN2D_firm_act).flatten().tolist()
CNN2D_firm_pred = np.array(CNN2D_firm_pred).flatten().tolist()
CNN2D_starch_act = np.array(CNN2D_starch_act).flatten().tolist()
CNN2D_starch_pred = np.array(CNN2D_starch_pred).flatten().tolist()

# Create DataFrames
CNN2D_brix_res = pd.DataFrame({
    'Actual': CNN2D_brix_act,
    'Predicted': CNN2D_brix_pred
})

CNN2D_firm_res = pd.DataFrame({
    'Actual': CNN2D_firm_act,
    'Predicted': CNN2D_firm_pred
})

CNN2D_starch_res = pd.DataFrame({
    'Actual': CNN2D_starch_act,
    'Predicted': CNN2D_starch_pred
})

# Optional: preview the results
print(CNN2D_brix_res.head())
print(CNN2D_firm_res.head())
print(CNN2D_starch_res.head())
```

```python
# Create actual vds predicted plots for each model and feature using sns
fig, axes = plt.subplots(1, 3, figsize=(11, 4))
sns.scatterplot(data=ViT_brix, x='Actual', y='Predicted', ax=axes[0])
axes[0].set_title('Brix')
sns.scatterplot(data=ViT_firm, x='Actual', y='Predicted', ax=axes[1])
axes[1].set_title('Firmness')
sns.scatterplot(data=ViT_starch, x='Actual', y='Predicted', ax=axes[2])
axes[2].set_title('Starch')
# add y=x line to each plot
for ax in axes:
    ax.plot([ax.get_xlim()[0], ax.get_xlim()[1]], [ax.get_ylim()[0], ax.get_ylim()[1]], 'r--')
plt.suptitle('ViT Model: Actual vs Predicted')
plt.tight_layout()
plt.show()

```
