```python
import pickle

model = "2D_CNN"
path = f"all_results_{model}_brix.pkl"

# Load list of dicts
with open(path, "rb") as f:
    results = pickle.load(f)

# Filter by timestamp prefix
filtered = [
    entry for entry in results
    if entry.get("timestamp", "").startswith("20260220_")
]

print(f"Kept {len(filtered)} entries out of {len(results)} total.")

with open(f"all_results_{model}_brix_filtered.pkl", "wb") as f:
    pickle.dump(filtered, f)
```
