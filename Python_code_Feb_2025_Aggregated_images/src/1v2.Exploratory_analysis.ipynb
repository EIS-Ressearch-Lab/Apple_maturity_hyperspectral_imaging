```python
import numpy as np
import pandas as pd
```

```python
file_path = '/home/ht21074/Auto_box_apples/Auto_box_apples/Source_folder/csv_files/All_data_csv/'

data_NZ2023 = pd.read_csv(f'{file_path}data_NZ2023.csv')
data_NZ2024 = pd.read_csv(f'{file_path}data_NZ2024.csv')
data_UK2024 = pd.read_csv(f'{file_path}data_UK2024.csv')

data = pd.concat([data_NZ2023, data_NZ2024, data_UK2024], ignore_index=True)
```

```python
print(data.head(5))
```

```python
# Plot brix vs firmness vs starch
import matplotlib.pyplot as plt

# plt.plot.figure(figsize=(10, 6))
plt.scatter(data['Brix'], data['Avg_firmness'], c=data['Starch'], cmap='viridis', alpha=0.5)
plt.colorbar(label='Starch')
plt.xlabel('Brix')
plt.ylabel('Firmness')
plt.title('Brix vs Firmness vs Starch')
plt.show()
```

```python
# Plot brix vs firmness vs starch facet by culitvar

import seaborn as sns
g = sns.FacetGrid(data, col='Cultivar', col_wrap=3, height=4)
g.map_dataframe(sns.scatterplot, x='Brix', y='Avg_firmness', hue='Starch', palette='viridis', alpha=0.5)
g.add_legend()
g.set_axis_labels('Brix', 'Firmness')
g.set_titles(col_template="{col_name}")
plt.show()
```

```python
# Plot brix vs firmness vs starch facet by culitvar with brix and firmness on y axes and starch on x axis
g = sns.FacetGrid(data, col='Cultivar', col_wrap=3, height=4)
g.map_dataframe(sns.scatterplot, x='Starch', y='Brix', hue='Avg_firmness', alpha=0.5)
g.add_legend()
g.set_axis_labels('Starch', 'Brix')
g.set_titles(col_template="{col_name}")
plt.show()
```

```python

g = sns.PairGrid(data[['Brix', 'Avg_firmness','Starch', 'Cultivar']], height=3, aspect=1, hue = 'Cultivar')
g.map(sns.scatterplot)
g.add_legend()
g.savefig('/home/ht21074/Auto_box_apples/Auto_box_apples/Source_folder/Generated_images/pairgrid_brix_firmness_starch.png', dpi=400)
g.savefig('/home/ht21074/Auto_box_apples/Auto_box_apples/Source_folder/Generated_images/pairgrid_brix_firmness_starch.jpg', dpi=400)
```
