## Creating training data for the Bayesian optimisation function
# Outputs are in '/home/ht21074/Auto_box_apples/Auto_box_apples/Source_folder/Training_data_50px/all_years'

```python
import numpy as np
import random
from sklearn.utils import shuffle
```

```python
img_size = 40

# X_train brix, firmness and starch and their encoders are the same
if img_size == 30 or img_size == 20:
    training_data_path = '/home/ht21074/Auto_box_apples/Auto_box_apples/Source_folder/Training_data_May2025/30px/all_years/'
elif img_size == 50 or img_size ==40:
    training_data_path = '/home/ht21074/Auto_box_apples/Auto_box_apples/Source_folder/Training_data_May2025/50px/all_years/'


X_train_brix = np.load(f'{training_data_path}X_train_all_years_Brix_shuffled.npy')
Y_train_brix = np.load(f'{training_data_path}Y_train_all_years_Brix_shuffled.npy')
X_train_firmness = np.load(f'{training_data_path}X_train_all_years_Firmness_shuffled.npy')
Y_train_firmness = np.load(f'{training_data_path}Y_train_all_years_Firmness_shuffled.npy')
X_train_starch = np.load(f'{training_data_path}X_train_all_years_Starch_shuffled.npy')
Y_train_starch = np.load(f'{training_data_path}Y_train_all_years_Starch_shuffled.npy')
Brix_encoder_shuffled  = np.load(f'{training_data_path}X_train_all_years_Brix_encoder_shuffled.npy')
Firmness_encoder_shuffled  = np.load(f'{training_data_path}X_train_all_years_Firmness_encoder_shuffled.npy')
Starch_encoder_shuffled  = np.load(f'{training_data_path}X_train_all_years_Starch_encoder_shuffled.npy')



```

```python
cultivars = ["Cox", "Braeburn", "Fuji", "Jazz", "Gala", "Golden Delicious"]

# Dictionary to store grouped data
cultivar_groups = {cultivar: [] for cultivar in cultivars}
cultivar_targets = {cultivar: [] for cultivar in cultivars}
encoders = {cultivar: [] for cultivar in cultivars}


cultivar_dict = {cultivar: [] for cultivar in cultivars}

for x, y, encoderx in zip(X_train_brix, Y_train_brix, Brix_encoder_shuffled):
    for cultivar in cultivars:
        if cultivar in x:  # Check if cultivar name is in file path
            cultivar_groups[cultivar].append(x)
            cultivar_targets[cultivar].append(y)
            encoders[cultivar].append(encoderx)
            break 

# Fix the random seed for reproducibility

random.seed(779)



```

```python
# Select 70% of each group
X_train_brix_optimisation = []
Y_train_brix_optimisation = []
X_encoder_brix_optimisation = []

for cultivar, files in cultivar_groups.items():
    num_samples = len(files)
    num_selected = max(1, int(0.70 * num_samples))  # Ensure at least one sample is selected
    selected_indices = random.sample(range(num_samples), num_selected)
    
    X_train_brix_optimisation.extend([files[i] for i in selected_indices])
    Y_train_brix_optimisation.extend([cultivar_targets[cultivar][i] for i in selected_indices])
    X_encoder_brix_optimisation.extend([encoders[cultivar][i] for i in selected_indices])

# Convert to NumPy arrays
X_train_brix_optimisation = np.array(X_train_brix_optimisation)
Y_train_brix_optimisation = np.array(Y_train_brix_optimisation)
X_encoder_brix_optimisation = np.array(X_encoder_brix_optimisation)


# Shuffle
X_train_brix_optimisation_shuffled, Y_train_brix_optimisation_shuffled, X_encoder_brix_optimisation = shuffle(X_train_brix_optimisation, Y_train_brix_optimisation, X_encoder_brix_optimisation, random_state=775)

opt_save_path = '/home/ht21074/Auto_box_apples/Auto_box_apples/Source_folder/Bayesian_optimisation_files/'
# Save as .npy files
np.save(f'{opt_save_path}X_train_brix_optimisation_May2025.npy', X_train_brix_optimisation_shuffled)
np.save(f'{opt_save_path}Y_train_brix_optimisation_May2025.npy', Y_train_brix_optimisation_shuffled)
np.save(f'{opt_save_path}X_encoder_brix_optimisation_May2025.npy', X_encoder_brix_optimisation)

# # # Side A
# np.save(f'{opt_save_path}X_train_brix_optimisation_May2025_SideA.npy', X_train_brix_optimisation_shuffled)
# np.save(f'{opt_save_path}Y_train_brix_optimisation_May2025_SideA.npy', Y_train_brix_optimisation_shuffled)
# np.save(f'{opt_save_path}X_encoder_brix_optimisation_May2025_SideA.npy', X_encoder_brix_optimisation)
```
